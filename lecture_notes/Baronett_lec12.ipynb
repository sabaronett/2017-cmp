```python
# MC intergral
import matplotlib.pyplot as plt
from math import sin
from random import random

def f(x):
    return sin(1/x/(2-x))**2

def MC1(N):
    count = 0
    for i in range(N):
        x = 2*random()
        y = random()
        if y<f(x):
            count += 1
    I = 2*count/N
    return I

def MC2(N):
    I = 0
    for i in range(N):
        x = 2*random()
        I += f(x)
    return I/N*2
```

```python
import numpy as np

N = 1000
trials = 100

mc1_arr = np.zeros(trials)
mc2_arr = np.zeros(trials)

for i in range(trials):
    mc1_arr[i] = MC1(N)
    mc2_arr[i] = MC2(N)

print("Variance in MC1:", np.var(mc1_arr))
print("Variance in MC2:", np.var(mc2_arr), "\n")
print("Std. error (RMS) in MC1:", np.std(mc1_arr))
print("Std. error (RMS) in MC2:", np.std(mc2_arr))
```

```python
plt.figure()
plt.plot(mc1_arr)
plt.plot(mc2_arr)
plt.show()
```
