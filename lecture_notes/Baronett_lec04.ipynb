```python
import matplotlib.pyplot as plt
from math import sin, cos
from numpy import linspace

x = []
y1 = []
y2 = []

for i in linspace(0,10,100):
    x.append(i)
    y1.append(sin(i))
    y2.append(cos(i))

plt.subplot(121)
plt.plot(x, y1, 'r', label='$\sin(x)$')
plt.legend(loc=1)

plt.subplot(122)
plt.plot(x, y2, 'o', label='$\cos(x)$')
plt.legend(loc=1)

plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.show()
```

```python
from math import sqrt,sin,pi
from numpy import empty

wavelength = 5.0
k = 2*pi/wavelength
xi0 = 1.0
separation = 10.0      # Separation of centers in cm
side = 100.0           # Side of the square in cm
points = 500           # Number of grid points along each side
spacing = side/points  # Spacing of points in cm


# Calculate the positions of the centers of the circles
x1 = side/4
y1 = side/4
x2 = side/2
y2 = side/2
x3 = (3/4)*side
y3 = (3/4)*side

# Make an array to store the heights
xi = empty([points,points],float)

# Calculate the values in the array
for i in range(points):
    y = spacing*i
    for j in range(points):
        x = spacing*j
        r1 = sqrt((x-x1)**2+(y-y1)**2)
        r2 = sqrt((x-x2)**2+(y-y2)**2)
        r3 = sqrt((x-x3)**2+(y-y3)**2)
        xi[i,j] = xi0*sin(k*r1) + xi0*sin(k*r2) + xi0*sin(k*r3)

# Make the plot
plt.imshow(xi,cmap="winter",origin="lower",extent=[0,side,0,side])
plt.colorbar()
#plt.gray()
plt.show()
```
