```python
from sklearn.datasets import load_iris
dataset = load_iris()
X = dataset.data
Y = dataset.target
print(X)
print(Y)
```

```python
from collections import OrderedDict
import pandas as pd

Iris_data = OrderedDict({'Sepal Length ': X[:,0],
             'Sepal Width': X[:,1],
             'Petal Length': X[:,2],
             'Petal Width': X[:,3],
             'Label': Y})

df = pd.DataFrame(Iris_data)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], c=Y, cmap=plt.cm.Set1,
            edgecolor='k')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.show();
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.scatter(X[:, 2], X[:, 3], c=Y, cmap=plt.cm.Set1,
            edgecolor='k')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.show();
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.scatter(X[:, 1], X[:, 0], c=Y, cmap=plt.cm.Set1,
            edgecolor='k')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.show();
```

```python
import matplotlib.pyplot as plt

label = ['Sepal Length','Sepal Width','Petal Length','Petal Width']


fig, axarr = plt.subplots(4, 4)
fig.set_size_inches(18, 18)

for i in range(4):
    x1 = X[:,i]
    for j in range(4):
        x2 = X[:,j]
        if i == j:
            axarr[i,j].text(0.2, 0.5, label[i],fontsize=20)
        else:
            axarr[i,j].scatter(x1, x2, c=Y, cmap=plt.cm.Set1,edgecolor='k')
            
plt.show()
```
