```python
import numpy as np

width = 100
height = 100
moves = 10000

x_start = int(width/2)
y_start = int(height/2)
pos = np.array([x_start, y_start])

def move_up(pos):
    pos[1] += 1
    
def move_down(pos):
    pos[1] -= 1

def move_left(pos):
    pos[0] -= 1

def move_right(pos):
    pos[0] += 1

print("Starting pos:", pos)
for i in range(moves):
    rand = np.random.randint(1, 5)
    if (rand == 1 and pos[1] != 99):
        move_up(pos)
    elif (rand == 2 and pos[1] != 0):
        move_down(pos)
    elif (rand == 3 and pos[0] != 0):
        move_left(pos)
    elif (rand == 4 and pos[0] != 99):
        move_right(pos)
    print("Pos after move", i, ":", pos)
```
