```python
# Author: Stanley A. Baronett
# Created on 7/29/19
# Last updated on 9/7/19
# 
# Homework 02: 3.5 Two Interacting Einstein Solids

import numpy as np
import matplotlib.pyplot as plt
from math import log, exp

def Factor(N):
    """
    Calculate and return the factorial of N.
    """
    N = int(N)
    total = 1
    if N < 0:
        print("Factorial error: N must be greater than 0")
    elif N > 1:
        for i in range(2, N+1):
            total *= i
    return total

def Multiplicity(N, q):
    """
    Calculate and return the multiplicity of N number
    of harmonic oscillators and q units of energy,
    using logarithms to accommodate large integer division.
    """
    return exp(log(Factor(q + N - 1)) - log(Factor(q)) - log(Factor(N - 1)))

def Plot_Omega(Na, Nb, q):
    """
    Plot the multiplicities, as a function of qa,
    of Na and Nb harmonic oscillators and q total
    units of energy.
    """
    qa_series = range(q + 1)
    Omega_A = []
    Omega_B = []
    Omega_tot = np.empty(1)
    
    for qa in qa_series:
        Omega_A.append(Multiplicity(Na, qa))
        Omega_B.append(Multiplicity(Nb, (q - qa)))
    
    Omega_A = np.asarray(Omega_A)
    Omega_B = np.asarray(Omega_B)
    Omega_tot = Omega_A * Omega_B
    
    plt.plot(qa_series, Omega_tot, linewidth=3, label='$\Omega(total)$')
    plt.plot(qa_series, Omega_A, "--", linewidth=2, label='$\Omega(A)$')
    plt.plot(qa_series, Omega_B, "-.", linewidth=2, label='$\Omega(B)$')
    plt.title('$N_A =$' + str(Na) + ', $N_B =$' + str(Nb) + ', $q =$' + str(q))
    plt.xlabel('$q_A$')
    plt.ylabel('$\Omega$')
    plt.legend(loc='best')
    plt.grid(color='black', alpha=0.2, linestyle='dashed', linewidth=0.5)
    plt.show()    
```

```python
Plot_Omega(3, 3, 2)
```

```python
Plot_Omega(300, 200, 100)
```

```python
Plot_Omega(3000, 2000, 100)
```
